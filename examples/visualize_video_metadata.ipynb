### This notebook walks through how to load a video stimulus and its corresponding metadata.

```python
import os
import json
from pprint import pprint

from IPython.display import Video
```

```python
dataset_root = "/data/vision/oliva/scratch/datasets/BOLDMomentsDataset" #"/your/path/to/BOLDMomentsDataset"

#load the metadata file
with open(os.path.join(dataset_root,"derivatives","stimuli_metadata","annotations.json"), 'r') as f:
    metadata = json.load(f)
```

```python
#display video metadata. The order of the dictionary fields does not matter.
#Read the metadata README and manuscript for an explanation of the metadata

#video ID to load
#videos 1-1000 are from the training set, and videos 1001-1102 are from the testing set
video_ID = 702
pprint(metadata[f"{video_ID:04}"])
```

```python
#display video mp4
Video(os.path.join(dataset_root, "derivatives", "stimuli_metadata", "mp4_h264", f"{video_ID:04}.mp4"), embed=True)
```
