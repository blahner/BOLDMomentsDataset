```python
import os
import json
from pprint import pprint
from PIL import Image
import glob as glob
import matplotlib.pyplot as plt
from IPython.display import Video
```

### This notebook walks through how to load a video stimulus and its corresponding metadata.

```python
dataset_root = "/your/path/to/BOLDMomentsDataset"

#load the metadata file
with open(os.path.join(dataset_root,"derivatives","stimuli_metadata","annotations.json"), 'r') as f:
    metadata = json.load(f)
```

```python
#display video metadata. The order of the dictionary fields does not matter.
#Read the metadata README and manuscript for an explanation of the metadata

#video ID to load
#videos 1-1000 are from the training set, and videos 1001-1102 are from the testing set
video_ID = 702
pprint(metadata[f"{video_ID:04}"])
```

```python
#display video mp4
Video(os.path.join(dataset_root, "derivatives", "stimuli_metadata", "mp4_h264", f"{video_ID:04}.mp4"), embed=True)
```

```python
#we also provide the video frames as jpg files. We load just the middle frame here but all frames are available.
#load middle frame
filepath = glob.glob(os.path.join(dataset_root, "derivatives","stimuli_metadata","frames_middle",f"{video_ID:04}_*.jpg"))
assert(len(filepath) == 1)
img = Image.open(filepath[0])

plt.imshow(img)
plt.axis('off')
plt.title(f"Middle frame of video {video_ID:04}", fontsize=12)
plt.tight_layout()
plt.show()
```
